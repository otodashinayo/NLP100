```python
from os import path
import pandas as pd
import torch
from torch import nn
base = "/content/drive/MyDrive/NLP100/ch08"

fp_train = "70/train.csv"
df_train = pd.read_csv(path.join(base, fp_train), index_col=0)
df_train.head()
```

```python
X_train = torch.tensor(df_train[[str(r) for r in range(300)]].values.astype("float32"))
y_train = torch.tensor(df_train["CATEGORY"].values.astype("int"))
print(X_train[:5])
print(y_train[:5])
```

```python
class Model(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(300, 4, bias=False)
    self.softmax = nn.Softmax(dim=1)

  def forward(self, x):
    y = self.linear(x)
    y = self.softmax(y)
    return y

model = Model()
print(model)
```

```python
loss_fn = nn.CrossEntropyLoss()
loss_1 = loss_fn(model(X_train)[0:1], y_train[0:1])
loss_1.backward()
grad_1 = model.linear.weight.grad

print("cross entropy loss")
print(loss_1)
print()
print("gradient")
print(grad_1)
print()
```

```python
loss_4 = loss_fn(model(X_train)[0:4], y_train[0:4])
loss_4.backward()
grad_4 = model.linear.weight.grad
print("cross entropy loss")
print(loss_4)
print()
print("gradient")
print(grad_4)
print()
```
